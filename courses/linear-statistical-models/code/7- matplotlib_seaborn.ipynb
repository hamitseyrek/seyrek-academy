```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("resources/factory_quality_regression.csv")
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x="quality_score", bins=10, kde=True)
plt.title("Kalite Skoru Dağılımı")
plt.xlabel("Kalite Skoru")
plt.ylabel("Frekans")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x="quality_score")
plt.title("Kalite Skoru Dağılımı")
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="runtime_hours", y="quality_score", hue = "line")
plt.title("Sensor Alert Skoru ile Kalite Skoru Arasındaki İlişki")
plt.xlabel("Sensor Alert Skoru")
plt.ylabel("Kalite Skoru")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="line", y="quality_score")
plt.title("Line'a Göre Kalite Skoru")
plt.xlabel("Line")
plt.ylabel("Kalite Skoru")
plt.show()
```
