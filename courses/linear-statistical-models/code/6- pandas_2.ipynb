```python
import pandas as pd

df = pd.read_csv("resources/factory_quality_regression.csv")

print(df.head())
print(df.info())
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
numeric_columns = [
    "sensor_alert_score",
    "uptime_rate",
    "maintenance_hours",
    "cycle_count",
    "quality_score",
]

for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)

df.isnull().sum()

```

```python
import numpy as np

df["quality_status"] = np.where(
    df["quality_score"] > 80,
    "good", 
    "bad")
df["asd"] = "nice"
df[["quality_score", "quality_status", "asd"]].head()
```

```python
df.sort_values("quality_score", ascending=True).head()
```

```python
df["risk_group"] = np.where(
    (df["uptime_rate"] < 70) & (df["runtime_hours"] < 6),
    "Yüksek Kalite Riski",
    "Takip",
)

df["risk_group"].value_counts()
```

```python
df.groupby("risk_group")[["quality_score"]].mean().reset_index()
```
