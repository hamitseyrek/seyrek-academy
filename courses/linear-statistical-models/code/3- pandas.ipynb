```python
import pandas as pd

print(pd.__version__)
```

```python
# Sözlük listesinden dataframe oluşturma
students = [
    {"name": "Ali", "age": 20, "city": "Ankara", "department": "RYZ", "gpa": 3.5, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Ayşe", "age": 21, "city": "İstanbul", "department": "RYZ", "gpa": 3.8, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Mehmet", "age": 22, "city": "İzmir", "department": "RYZ", "gpa": 3.2, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Zeynep", "age": 23, "city": "İstanbul", "department": "RYZ", "gpa": 3.9, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Ahmet", "age": 24, "city": "İzmir", "department": "RYZ", "gpa": 3.7, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Ayşe", "age": 25, "city": "İstanbul", "department": "RYZ", "gpa": 3.6, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Ali", "age": 26, "city": "Ankara", "department": "RYZ", "gpa": 3.4, "courses": ["Matematik", "Fizik", "Kimya"]},
]

df = pd.DataFrame(students)

print(df)



```

```python
print(df.head())

print(df.tail())

print(df.info())

print(df.describe())
```

```python
# Sütunları seçme => mean, min, max, sum, count, median, std

print(df["gpa"])
print(df["gpa"].mean())
print(df["gpa"].min())
print(df["gpa"].max())
print(df["gpa"].sum())
print(df["gpa"].count())
print(df["gpa"].median())
print(df["gpa"].std())




```

```python
daily_data = {
    "day": ["Pzt", "Sal", "Çar", "Per", "Cum", "Cmt", "Pzt"],
    "steps": [3000, 4000, 2000, 5000, 4000, 3000, 2000]
}

df_daily = pd.DataFrame(daily_data)

print(df_daily)
```

```python
sum_of_week = df_daily["steps"].mean()

print("Ortalama:", int(sum_of_week))


camp_day = df_daily["steps"].max()

print("En fazla adım:", camp_day)
```
