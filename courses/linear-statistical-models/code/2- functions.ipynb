Functions

```python
def average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

scores = [60,70,80,90]
average(scores)
```

```python
def mini(numbers):
    min = numbers[0]
    for i in numbers:
        if i < min:
            min = i
    return min

mini(scores)
print("En küçük sayı: ", mini(scores))
```

Q: Bir diziyi parametre olarak alıp dizideki elemanların toplamını dizinin ortalamasına bölen bir fonksiyon yazın.

```python
notes = [60,70,80,90, 100]

def complex(notes):
    total = sum(notes)
    average = total / len(notes)
    return total / average

complex(notes)
```

Q2: Bir sayı parametresi alıp faktöriyel hesaplayan fonksiyonu yazınız.

```python
def factorial(number):
    sum = 1
    for i in range(1, number + 1):
        sum *= i
    return sum

factorial(5)

def factorial2(number):
    if number == 0:
        return 1
    else:
        return number * factorial2(number - 1)

factorial2(5)

```
