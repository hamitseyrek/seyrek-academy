```python
# bu bir yorum satırıdır

notes = [70, 60, 50, 40, 30]

toplam = 0

for aa in notes:
    toplam += aa

print(toplam)

```

```python
notes = [70, 60, 50, 40, 30]

sum = sum(notes)
count = len(notes)
average = sum / count
minimum = min(notes)
maximum = max(notes)

print("Toplam:",sum)
print("Uzunluk:",count)
print("Ortalama:",average)
print("Minimum:",minimum)
print("Maksimum:",maximum)



```

```python
notes = [70, 60, 50, 40, 30]

min = notes[0]

for note in notes:
    if note < min:
        min = note

print(min)


```

```python
average = 60

if average >= 90:
    status = "AA"
elif average >= 85:
    status = "BA"
elif average >= 80:
    status = "BB"
elif average >= 75:
    status = "CB"
elif average >= 70:
    status = "CC"
elif average >= 65:
    status = "DC"
elif average >= 60:
    status = "DD"
else:
    status = "FF"
print(status)

```

```python
# Dictionary / Sözlük
# Key-Value pair

student = {
    "name": "Ali",
    "age": 20,
    "city": "Ankara",
    "department": "RYZ",
    "gpa": 3.5,
    "courses": ["Matematik", "Fizik", "Kimya"]
}

print(student)
print(student["name"])
print(student["age"])
print(student["city"])
print(student["department"])
print(student["gpa"])
print(student["courses"])


```

```python
#dictionary comprehension

students = [
    {"name": "Ali", "age": 20, "city": "Ankara", "department": "RYZ", "gpa": 3.5, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Ayşe", "age": 21, "city": "İstanbul", "department": "RYZ", "gpa": 3.8, "courses": ["Matematik", "Fizik", "Kimya"]},
    {"name": "Mehmet", "age": 22, "city": "İzmir", "department": "RYZ", "gpa": 3.2, "courses": ["Matematik", "Fizik", "Kimya"]},
]



for student in students:
    print(student["name"], "-> Ortalama:", student["gpa"])

    
```
